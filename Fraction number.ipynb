```python
class Fraction:
    
    def __init__(self,n,d):
        self.num = n
        self.din = d
        
    def __str__(self):
        return "{}/{}".format(self.num,self.din)
    
    def __add__(self,other):
        temp_num = self.num * other.din + self.din* other.num
        #print("temp num: ",temp_num)
        temp_din = self.din*other.din
        #print("temp din: ",temp_din)
        
        factors_num = []
        for i in list(range(1,temp_num+1)):
            if temp_num%i==0:
                factors_num.append(i)
        #print("factores of numeraror are: ",factors_num)
        factors_din = []
        for j in list(range(1,temp_din+1)):
            if temp_din%j==0:
                factors_din.append(j)
        #print("factors of dinominator: ",factors_din)
        common_fact = []
        for i in factors_num:
            if i in factors_din:
                common_fact.append(i)
        #print("common factors are: ",common_fact)
        divisor = max(common_fact)
        #print("divisor",divisor)
        num = temp_num//divisor
        din = temp_din//divisor
        
        return "{}/{}".format(num,din)
    
    def __sub__(self,other):
        temp_num = self.num * other.din - self.din* other.num
        #print("temp num: ",temp_num)
        temp_din = self.din*other.din
        #print("temp din: ",temp_din)
        
        factors_num = []
        for i in list(range(1,temp_num+1)):
            if temp_num%i==0:
                factors_num.append(i)
        #print("factores of numeraror are: ",factors_num)
        factors_din = []
        for j in list(range(1,temp_din+1)):
            if temp_din%j==0:
                factors_din.append(j)
        #print("factors of dinominator: ",factors_din)
        common_fact = []
        for i in factors_num:
            if i in factors_din:
                common_fact.append(i)
        #print("common factors are: ",common_fact)
        divisor = max(common_fact)
        #print("divisor",divisor)
        num = temp_num//divisor
        din = temp_din//divisor
        
        return "{}/{}".format(num,din)
    
    def __mul__(self,other):
        temp_num = self.num * other.num
        #print("temp num: ",temp_num)
        temp_din = self.din*other.din
        #print("temp din: ",temp_din)
        
        factors_num = []
        for i in list(range(1,temp_num+1)):
            if temp_num%i==0:
                factors_num.append(i)
        #print("factores of numeraror are: ",factors_num)
        factors_din = []
        for j in list(range(1,temp_din+1)):
            if temp_din%j==0:
                factors_din.append(j)
        #print("factors of dinominator: ",factors_din)
        common_fact = []
        for i in factors_num:
            if i in factors_din:
                common_fact.append(i)
        #print("common factors are: ",common_fact)
        divisor = max(common_fact)
        #print("divisor",divisor)
        num = temp_num//divisor
        din = temp_din//divisor
        
        return "{}/{}".format(num,din)
    
    def __truediv__(self,other):
        temp_num = self.num * other.din
        #print("temp num: ",temp_num)
        temp_din = self.din*other.num
        #print("temp din: ",temp_din)
        
        factors_num = []
        for i in list(range(1,temp_num+1)):
            if temp_num%i==0:
                factors_num.append(i)
        #print("factores of numeraror are: ",factors_num)
        factors_din = []
        for j in list(range(1,temp_din+1)):
            if temp_din%j==0:
                factors_din.append(j)
        #print("factors of dinominator: ",factors_din)
        common_fact = []
        for i in factors_num:
            if i in factors_din:
                common_fact.append(i)
        #print("common factors are: ",common_fact)
        divisor = max(common_fact)
        #print("divisor",divisor)
        num = temp_num//divisor
        din = temp_din//divisor
        
        return "{}/{}".format(num,din)      
```
